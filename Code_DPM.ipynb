```python
dates = filtered_columnss.columns.tolist()  
prices = filtered_columnss.iloc[0].tolist()  

volumes = []
for date in dates:
    if date in filtered_columnss1.columns:  
        volume = filtered_columnss1[date].tolist()[0]  
        volumes.append(volume)
    else:
        volumes.append(None)  

fig = go.Figure()

fig.add_trace(go.Scatter(x=dates, y=prices, name="Giá", line=dict(color=color1)))
fig.add_trace(go.Scatter(x=dates, y=pd.Series(prices).rolling(window=50).mean(), name="MA 50"))
fig.add_trace(go.Scatter(x=dates, y=pd.Series(prices).rolling(window=100).mean(), name="MA 100"))
fig.add_trace(go.Scatter(x=dates, y=pd.Series(prices).rolling(window=200).mean(), name="MA 200"))

fig.add_trace(go.Bar(x=dates, y=volumes, name="Khối lượng", marker_color=color))

fig.update_layout(title="CHỈ BÁO MOVING AVERAGE",
                  xaxis_title="",
                  barmode='overlay',
                  xaxis=dict(
                  tickvals=[], 
                  ticktext=[] 
                  )
)

fig.show()
```

```python

window = 20
std_dev_multiplier = 2

sma = pd.Series(prices).rolling(window=window).mean()  
std_dev = pd.Series(prices).rolling(window=window).std()  

upper_band = sma + std_dev_multiplier * std_dev  
lower_band = sma - std_dev_multiplier * std_dev  

fig = go.Figure()

fig.add_trace(go.Scatter(x=dates, y=prices, name="Giá", line=dict(color=color1)))

fig.add_trace(go.Scatter(x=dates, y=upper_band, name="Dải trên"))
fig.add_trace(go.Scatter(x=dates, y=lower_band, name="Dải dưới"))

fig.add_trace(go.Bar(x=dates, y=volumes, name="Khối lượng", marker_color=color))

fig.update_layout(title="CHỈ BÁO BOLLINGER BANDS",
                  xaxis_title="",
                  barmode='overlay',
                  xaxis=dict(
                      tickvals=[],  
                      ticktext=[]  
                  )
                  )

fig.show()
```

```python

exp12 = pd.Series(prices).ewm(span=12, adjust=False).mean()
exp26 = pd.Series(prices).ewm(span=26, adjust=False).mean()
macd_line = exp12 - exp26
signal_line = macd_line.ewm(span=9, adjust=False).mean()
histogram = macd_line - signal_line

fig = go.Figure()

fig.add_trace(go.Scatter(x=dates, y=macd_line, name="Đường MACD", line=dict(color="blue")))
fig.add_trace(go.Scatter(x=dates, y=signal_line, name="Đường tín hiệu", line=dict(color="red")))
fig.add_trace(go.Bar(x=dates, y=histogram, name="Biểu đồ MACD", marker=dict(color="purple")))

fig.update_layout(title="CHỈ BÁO MOVING AVERAGE CONVERGENCE DIVERGENCE",
    xaxis=dict(
        showticklabels=False  
    ),
    yaxis=dict(
    )
)

fig.show()
```

```python
delta = pd.Series(prices).diff()
gain = delta.apply(lambda x: x if x > 0 else 0)
loss = delta.apply(lambda x: abs(x) if x < 0 else 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
rsi_values = 100 - (100 / (1 + rs))

fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=rsi_values, name="RSI 14"))
fig.add_shape(
    type="line",
    x0=dates[0],
    y0=30,
    x1=dates[-1],
    y1=30,
    line=dict(color="red", width=1, dash="dash")
)
fig.add_shape(
    type="line",
    x0=dates[0],
    y0=70,
    x1=dates[-1],
    y1=70,
    line=dict(color="green", width=1, dash="dash")
)
fig.update_layout(
    xaxis=dict(
        title="",
        ticktext=[],  
        tickvals=[]  
    ),
    yaxis=dict(
        title="",
        range=[0, 100]
    )
)
fig.update_layout(title="CHỈ BÁO RELATIVE STRENGTH INDEX",
    xaxis=dict(
        showticklabels=False  
    ),
    yaxis=dict(
    )
)

fig.show()
```
